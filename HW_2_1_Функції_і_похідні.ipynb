1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import matplotlib.pyplot as plt
import sympy as sp
import numpy as np
from sympy import symbols, diff, sin, cos, ln, lambdify
```

```python
def find_derivative(function_param):
  x = sp.symbols('x')
  function = function_param
  derivative = sp.diff(function, x)
  print(f'Похідна ф-ії: {derivative}')

find_derivative(function_param=10*x-5)
```

```python
x_values = np.linspace(0,10,100)
f_lin = 10*x_values-5
f_der = np.full_like(f_lin, 10)
plt.figure(figsize=(10, 5))
plt.plot(x_values, f_lin, label='f(x) = 10*x-5', color='blue')
plt.plot(x_values, f_der, label="f'(x) = 10", color='red')
plt.title('Значення функції vs значення похідної')
plt.legend()
plt.show()
```

**Висновок**:
1. Значення оцінки лінійно збільшується від збільшення кількості годин.

2. Але швидкість зміни оцінки є сталою і не змінюється в залежності від кількості годин.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
f = sin(x)
find_derivative(function_param=f)
```

```python
def viz(x_param, function_param):
  x = sp.symbols('x')
  function = function_param
  derivative = sp.diff(function, x)
  x_values = np.linspace(0,x_param,100)


  f_lin = lambdify(x,function,'numpy')
  f_der = lambdify(x,derivative,'numpy')

  f_values = f_lin(x_values)
  f_derivative_values = f_der(x_values)

  plt.figure(figsize=(10, 5))
  plt.plot(x_values, f_values, label=f'f(x)={function_param}', color='blue')
  plt.plot(x_values, f_derivative_values, label=f"f'(x) ={derivative}", color='red')
  plt.title('Значення функції vs значення похідної')
  plt.legend()
  plt.show()
```

```python
viz(x_param = 2* np.pi, function_param = f)
```

```python
viz(x_param = 10, function_param = ln(x))
```
